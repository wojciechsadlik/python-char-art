```python
from PIL import Image, ImageFont
import matplotlib.pyplot as plt
from img_processing import preprocess_img, DITHER_MODES
from generate_char_palette import get_asciis, get_ansi_256_colors_grayscale_set
from ansi_colorizer import AnsiColorizer, reset_code
from mono_char_art_converter import MonoCharArtConverter
```

```python
IMG_PATH = "imgs/peppers.tiff"
FONT_PATH = "fonts/Inconsolata-Regular.ttf"

img = Image.open(IMG_PATH).convert("RGB")
font = ImageFont.truetype(FONT_PATH, 32)
```

```python
proc_img = preprocess_img(img=img, scale_factor=0.1, sharpness=2, grayscale=True)
plt.imshow(proc_img, cmap='gray', vmin=0, vmax=255, interpolation='none')
plt.show()

converter = MonoCharArtConverter(
    char_set=get_asciis(),
    font=font,
    general_mapping_palette_shape=(12,12)
)

ascii_arr = converter.convert(proc_img)
for i in range(len(ascii_arr)):
    for j in range(len(ascii_arr[i])):
        print(ascii_arr[i][j], sep='', end='')
    print('\n', end='')
```

```python
proc_img = preprocess_img(img=img, scale_factor=0.2, sharpness=2, grayscale=True)
plt.imshow(proc_img, cmap='gray', vmin=0, vmax=255, interpolation='none')
plt.show()

converter = MonoCharArtConverter(
    char_set=get_asciis(),
    font=font,
    general_mapping_palette_shape=(12,12),
    detailed_mapping_wh=(2,4)
)

ascii_arr = converter.convert(proc_img)
for i in range(len(ascii_arr)):
    for j in range(len(ascii_arr[i])):
        print(ascii_arr[i][j], sep='', end='')
    print('\n', end='')
```

```python
proc_img = preprocess_img(img=img, scale_factor=0.1, sharpness=2, grayscale=True)
plt.imshow(proc_img, cmap='gray', vmin=0, vmax=255, interpolation='none')
plt.show()

converter = MonoCharArtConverter(
    char_set=get_ansi_256_colors_grayscale_set(get_asciis(), 8),
    font=font,
    general_mapping_palette_shape=(16,16),
    normalize_palette=False
)

ascii_arr = converter.convert(proc_img)
for i in range(len(ascii_arr)):
    for j in range(len(ascii_arr[i])):
        print(ascii_arr[i][j], sep='', end='')
    print('\n', end='')
```

```python
proc_img = preprocess_img(img=img, scale_factor=0.2, sharpness=2, grayscale=False)
plt.imshow(proc_img, cmap='gray', vmin=0, vmax=255, interpolation='none')
plt.show()

colorize_settings = AnsiColorizer(
    colored_fg=True,
    colored_bg=True,
    fg_brightness_scale=1.0,
    bg_brightness_scale=0.4,
    use_ansi_256_colors=False
)
converter = MonoCharArtConverter(
    char_set=get_asciis(),
    font=font,
    general_mapping_palette_shape=(12,12),
    detailed_mapping_wh=(2,4),
    colorize_settings=colorize_settings
)
ascii_arr = converter.convert(proc_img)
for i in range(len(ascii_arr)):
    for j in range(len(ascii_arr[i])):
        print(ascii_arr[i][j], sep='', end='')
    print(reset_code() + '\n', end='')
```
