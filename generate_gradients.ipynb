```python
from PIL import Image, ImageOps
```

```python
l_grad = ImageOps.scale(Image.linear_gradient("L"), 2)

l_grad.save("imgs/lin_grad_ud.bmp", "BMP")

l_grad
```

```python
l_grad.rotate(90).save("imgs/lin_grad_lr.bmp", "BMP")

l_grad.rotate(90)
```

```python
l_grad.rotate(180).save("imgs/lin_grad_du.bmp", "BMP")

l_grad.rotate(180)
```

```python
r_grad = ImageOps.scale(Image.radial_gradient("L"), 2)

r_grad.save("imgs/rad_grad.bmp", "BMP")

r_grad
```

```python
ir_grad = ImageOps.invert(r_grad)

ir_grad.save("imgs/irad_grad.bmp", "BMP")

ir_grad
```
